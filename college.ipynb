```python
import pandas as pd
```

```python
df= pd.read_csv('college.csv')
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
df['target']=lbl.fit_transform(df['Private'])
```

```python
df['target'].dtype
```

```python
df.drop(['Private'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
y=df[['target']]
```

```python
x=df[['Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc', 'F.Undergrad',
       'P.Undergrad', 'Outstate', 'Room.Board', 'Books', 'Personal', 'PhD',
       'Terminal', 'S.F.Ratio', 'perc.alumni', 'Expend', 'Grad.Rate']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8 ,random_state=25)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
svcmodel=SVC()
svcmodel.fit(x_train,y_train)
```

```python
pred=svcmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(confusion_matrix(y_test, pred))
print()
print(classification_report(y_test, pred))
print()
print(accuracy_score(y_test, pred))
```

```python
X=df[['Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc', 'F.Undergrad',
       'P.Undergrad', 'Outstate', 'Room.Board', 'Books', 'Personal', 'PhD',
       'Terminal', 'S.F.Ratio', 'perc.alumni', 'Expend', 'Grad.Rate']]
```

```python
from sklearn.preprocessing import StandardScaler
sds=StandardScaler()
sds.fit(X)
X=sds.transform(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.8 ,random_state=58)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
svcmodel=SVC()
svcmodel.fit(x_train,y_train)
```

```python
pred=svcmodel.predict(x_test)
```

```python
print(confusion_matrix(y_test, pred))
print()
print(classification_report(y_test, pred))
print()
print(accuracy_score(y_test, pred))
```

```python
parameters = {'gamma':(1,0.1,0.01,.001), 'C':[0.1,1,10,100]}
from sklearn.model_selection import GridSearchCV
gs=GridSearchCV(svcmodel,parameters,refit=True,verbose=2)
```

```python
gs.fit(x_train,y_train)
```

```python
grid_prediction=gs.predict(x_test)
```

```python
print("Accuracy score:")
print(accuracy_score(y_test,grid_prediction))
```
