```python
train_data=train_data[train_data.y<150]
print(train_data.shape)
```

# dimentionality reduction

```python
#feaures
train_x=train_data.drop(['ID','y'],axis=1)
test_x=test_data.drop(['ID'],axis=1)
#output
train_y=train_data['y']
train_y=train_y.values
#matching cols
matching=train_x.columns.intersection(test_x.columns)
matc_list=matching.tolist()

test_x=test_x[matc_list]
train_x=train_x[matc_list]

print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.3, random_state=101)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
scaler = StandardScaler()
```

```python
pca = PCA(n_components=6)
```

```python
train_sc_x=scaler.fit_transform(x_train)
#train_sc_y=scaler.fit_transform(y_train)
print(train_sc_x.shape)
#print(train_sc_y.shape)
```
